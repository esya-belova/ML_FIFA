# Настройка гиперпараметров

```python
from tensorflow.keras import regularizers
from tensorboard.plugins.hparams import api as hp
```

```python
def build_model(n_hidden=2, n_neurons=30, learning_rate=3e-3, input_shape=(23,),activity_regularizer=regularizers.l2(1e-5)): 
    activity_regularizer=regularizers.l2(activity_regularizer)
    model = keras.models.Sequential()
    model.add(keras.layers.Flatten(input_shape=(23,)))
    for layer in range(n_hidden):
        model.add(keras.layers.Dense(n_neurons, activation="relu",activity_regularizer=activity_regularizer))
    model.add(keras.layers.Dense(4, activation="softmax"))
    optimizer = keras.optimizers.SGD(lr=0.01)
    model.compile(loss="categorical_crossentropy", optimizer=optimizer, metrics=["accuracy"])
    return model
```

```python
keras_Class = keras.wrappers.scikit_learn.KerasClassifier(build_model)
```

```python
from scipy.stats import reciprocal
from sklearn.model_selection import RandomizedSearchCV


param_distribs = {
   
    "n_neurons": [5,15,25,40,60,80,100,200],
    "activity_regularizer":[0.001, 0.01, 0.1, 0.5, 1, 5, 10, 50, 100],
}


rnd_search_cv = RandomizedSearchCV(keras_Class, param_distribs, n_iter=10, cv=3, verbose=2)
history=rnd_search_cv.fit(X_train, y_train, epochs=10)
```

```python
from tensorflow.keras import regularizers
from tensorboard.plugins.hparams import api as hp
```

```python
import shutil
#shutil.rmtree("./logs/")
```

```python
HP_N_NEURONS = hp.HParam('n_neurons', hp.Discrete([5,15,25,40,60,80,100,200]))
HP_ACT_REG = hp.HParam('activity_regularizer', hp.Discrete([0, 1, 5, 10, 50]))

METRIC_ACCURACY = 'accuracy'
METRIC_LOSS='loss'
with tf.summary.create_file_writer('logs/hparam_tuning').as_default():
    hp.hparams_config(
        hparams=[HP_N_NEURONS, HP_ACT_REG],
        metrics=[hp.Metric(METRIC_ACCURACY, display_name='Accuracy'),
                hp.Metric(METRIC_LOSS, display_name='Loss')],
      )
```

```python
def train_test_model(hparams):
    activity_regularizer=regularizers.l2(hparams[HP_ACT_REG])
    model = tf.keras.models.Sequential([
        tf.keras.layers.Flatten(input_shape=(23,)),
        tf.keras.layers.Dense(hparams[HP_N_NEURONS], activation=tf.nn.relu, activity_regularizer=activity_regularizer),
        tf.keras.layers.Dense(4, activation=tf.nn.softmax),
      ])
    model.compile(
        optimizer=keras.optimizers.SGD(lr=0.01),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy','sparse_categorical_crossentropy'],
      )

    
    history=model.fit(X_train, y_train, epochs=10) 
    _,accuracy,loss = model.evaluate(X_test, y_test)
    return loss,accuracy
    
```

```python
def run(run_dir, hparams):
    with tf.summary.create_file_writer(run_dir).as_default():
        hp.hparams(hparams)  # record the values used in this trial
        loss,accuracy = train_test_model(hparams)
        tf.summary.scalar(METRIC_LOSS, loss, step=1)
        tf.summary.scalar(METRIC_ACCURACY, accuracy, step=2)
```

```python
session_num = 0

for num_neuron in HP_N_NEURONS.domain.values:
    for actreg in HP_ACT_REG.domain.values:
        hparams = {
          HP_N_NEURONS: num_neuron,
          HP_ACT_REG: actreg,
          }
        run_name = "run-%d" % session_num
        print('--- Starting trial: %s' % run_name)
        print({h.name: hparams[h] for h in hparams})
        run('logs/hparam_tuning/' + run_name, hparams)
        session_num += 1
```

Исследование проведено с помощью подброки гиперпараметров (Random Search) посредством фреймворка tensorflow и инструмента вывода графиков tensorboard. Благодаря этому, можно сделать вывод, что максимальная точность достигается при n=40 и alpha=0.01 (где n-количество нейронов на слой, а alpha-параметр регуляризации). При этом alpha не должен быть слишком большим, при alpha>1 точность начинает ухудшаться, а n не должен быть слишком маленьким, при n<15 результат также будет падать. 
Обучение проходило на 10 эпохах, данное количество эпох является оптимальным, так как при бОльшем количестве эпох наблюдается переобучение.
